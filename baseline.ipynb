```python
import templater3
import random
```

```python
def template_line(line):
    page, code = line.split('\t')

    template = templater3.Templater(min_block_size=2)
    for line in code.replace('\\n', '\n').splitlines():
        template.learn(line)
        print(line)
    print(template._template)
    
line = open('simplewiki-20211120-lists-1k.tsv').readlines()[0]
template_line(line)
```

```python
def generate_numbers(size):
    numbers = random.sample(range(1001), size)
    return numbers

chosen = generate_numbers(30)
```

```python
for number in chosen:
    line = open('simplewiki-20211120-lists-1k.tsv').readlines()[number]
    template_line(line)
    print('\n------------------------------------------------------------------------------------------------------------------------------------- \n')
```
